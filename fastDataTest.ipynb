```python
# extract the uv and vis and the scan number
# get flag

fname_ms  = 'data/20240504_194310_73MHz_calibrated_selfcalibrated_sun_only.ms'

tb = casatools.table()
tb.open(fname_ms)
uvw = tb.getcol('UVW')
data = tb.getcol('DATA')
scan = tb.getcol('SCAN_NUMBER')
flag = tb.getcol('FLAG')
tb.close()
myTB = casatools.table()
myTB.open(fname_ms + '/SPECTRAL_WINDOW')
# Read the REF_FREQUENCY and NUM_CHAN columns into Python arrays
chan_freqs = myTB.getcol('CHAN_FREQ')
myTB.close()
wavelen = 3e8 / np.mean(chan_freqs)
tb.open(fname_ms + '/FIELD')
# Get the reference RA/DEC for the first field
ref_ra_dec = tb.getcol('PHASE_DIR').squeeze()
tb.close()
# convert the uvw to wavelength lambda

uvw = uvw / wavelen

chan_id  =  6
stokes_I = np.mean(0.5 * (data[0,5:7,:] + data[-1,5:7,:]), axis=0).squeeze()
stokes_I = 0.5 * (data[0,chan_id,:] + data[-1,chan_id,:])
flag_I = np.any(np.any(flag, axis=0),axis=0)
```

```python
uvw.shape
```

```python
process_idx = 1

# how many scan
scan_unique = np.unique(scan)
#for this_scan in scan_unique:
#    ind = np.where(scan == this_scan)[0]
    #plt.plot(uvw[0,ind], uvw[1,ind], 'o')
ind = np.where(scan == process_idx)[0]

ref_proc = ref_ra_dec[:,process_idx]

uvw_this_scan = uvw[:,ind]
stokes_I_this_scan = stokes_I[ind]
uvw_this_scan_goodvis = uvw[:,ind][:,~flag_I[ind]]
stokes_I_this_scan_goodvis = stokes_I[ind][~flag_I[ind]]

idx_gt_3lambda = np.where(np.sqrt(np.sum(uvw_this_scan_goodvis**2, axis=0)) > 30)[0]
uvw_this_scan_goodvis = uvw_this_scan_goodvis[:,idx_gt_3lambda]
stokes_I_this_scan_goodvis = stokes_I_this_scan_goodvis[idx_gt_3lambda]
```

```python
np.savez('data/uvw_stokesI.npz', uvw = uvw, scan=scan, flag_I=flag_I, stokes_I=stokes_I)
```

```python
# plot the amplitude and phase
amp = np.abs(stokes_I_this_scan_goodvis)
phase = np.angle(stokes_I_this_scan_goodvis)
fig, axs = plt.subplots(1,2, figsize=(10,4.5))

axs = axs.ravel()
axs[0].scatter(uvw_this_scan_goodvis[0], uvw_this_scan_goodvis[1], c=(amp), s=15)
#axs[0].set_aspect('equal', 'box')
axs[1].scatter(uvw_this_scan_goodvis[0], uvw_this_scan_goodvis[1], c=np.cos(phase), s=15, cmap='inferno')
#axs[1].set_aspect('equal', 'box')
```
